```python
import nltk
import heapq

# Import stopwords (common words like "is", "the", etc.) and tokenizers
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize


# Sample input text (this is what we want to summarize)
text = """India is a developing country. It has a large population. 
Technology is growing fast in India. Many people are using the internet. 
Education is also improving day by day. India is becoming digital."""

# -------------------- Preprocessing --------------------

# Load English stopwords (words we want to ignore)
stop_words = set(stopwords.words('english'))

# Break the text into individual words
words = word_tokenize(text)

# Dictionary to store frequency of each word
freq = {}

# Count how many times each important word appears
for word in words:
    word = word.lower()  # Convert to lowercase for consistency
    
    # Check if word is alphanumeric and not a stopword
    if word.isalnum() and word not in stop_words:
        freq[word] = freq.get(word, 0) + 1  # Increase count

# Find the maximum frequency (used for normalization)
max_freq = max(freq.values())

# Normalize frequencies (scale values between 0 and 1)
for word in freq:
    freq[word] = freq[word] / max_freq

# -------------------- Sentence Scoring --------------------

# Split text into sentences
sentences = sent_tokenize(text)

# Dictionary to store score of each sentence
sent_score = {}

# Calculate score for each sentence
for sent in sentences:
    for word in word_tokenize(sent.lower()):
        
        # If the word exists in frequency dictionary, add its score
        if word in freq:
            sent_score[sent] = sent_score.get(sent, 0) + freq[word]

# -------------------- Summary Generation --------------------

# Select top 2 sentences with highest scores
summary = heapq.nlargest(2, sent_score, key=sent_score.get)

# -------------------- Output --------------------

print("Original Text:\n", text)

# Join selected sentences to form final summary
print("\nSummary:\n", " ".join(summary))
```
